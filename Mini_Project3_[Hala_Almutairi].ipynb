##6.Code

###1 - Import and Initial Inspection

```python
import pandas as pd
import re
df_Airplane_Crashes =pd.read_csv("/content/drive/MyDrive/Phase2/Week3/MiniProject3/Airplane_Crashes (1).csv")
df_Airplane_Crashes.head(5)
```

```python
df_Airplane_Crashes.info()
```

```python
df_Airplane_Crashes.isnull().sum()
```

###2 - Drop Irrelevant Columns

```python
df_Airplane_Crashes.drop(columns=['Flight #', 'Registration', 'cn/In', 'Ground','Operator','Route','Type','Location'], inplace = True)
df_Airplane_Crashes
df_Airplane_Crashes.isnull().sum()
```

###3 - Handle Missing Data

```python
#1. Time (2219 missing)
df_Airplane_Crashes['Time'] = df_Airplane_Crashes['Time'].fillna('Unknown')
#2. Aboard (22 missing)
df_Airplane_Crashes = df_Airplane_Crashes.dropna(subset=['Aboard'])
#3. Fatalities (12 missing)
df_Airplane_Crashes = df_Airplane_Crashes.dropna(subset=['Fatalities'])
#4. Summary (390 missing)
df_Airplane_Crashes['Summary'] = df_Airplane_Crashes['Summary'].fillna('')
df_Airplane_Crashes.isnull().sum()
df_Airplane_Crashes
```

###4 - Clean and Standardize Key Columns

####4.1 - Date

#####4.1.1 Extract Year and Month

```python
df_Airplane_Crashes['Date'] = pd.to_datetime(df_Airplane_Crashes['Date'], format='%m/%d/%Y')

df_Airplane_Crashes['Year'] = df_Airplane_Crashes['Date'].dt.year
df_Airplane_Crashes['Month'] = df_Airplane_Crashes['Date'].dt.month
df_Airplane_Crashes.head(2)
```

#####4.1.2 Define and Extract Season

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

df_Airplane_Crashes['Season'] = df_Airplane_Crashes['Month'].apply(get_season)
df_Airplane_Crashes['Season'] = df_Airplane_Crashes['Month'].apply(get_season)
df_Airplane_Crashes.head(2)
```

####4.2 -Time

```python
df_Airplane_Crashes['Time'] = df_Airplane_Crashes['Time'].astype(str).apply(lambda x: re.sub(r'[^0-9:]', '', x) if x != 'Unknown' else 'Unknown')

df_Airplane_Crashes['Time'] = df_Airplane_Crashes['Time'].where(
    df_Airplane_Crashes['Time'].str.match(r'^\d{1,2}:\d{2}$'),
    'Unknown'
)

df_Airplane_Crashes.head(20)
```

#####4.2.1 Extract Hour from Time and Categorize Hour into TimeOfDay

```python
def extract_hour(time_str):
    if time_str == 'Unknown' or pd.isna(time_str):
        return None
    else:
        return int(time_str.split(':')[0])

def get_time_of_day(hour):
    if hour is None:
        return 'Unknown'
    elif 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df_Airplane_Crashes['TimeOfDay'] = df_Airplane_Crashes['Time'].apply(extract_hour).apply(get_time_of_day)
df_Airplane_Crashes
```

####4.3 -Aboard And Fatalities

```python
df_Airplane_Crashes['Aboard'] = df_Airplane_Crashes['Aboard'].astype(str).str.replace(r'\D', '', regex=True).astype(int)
df_Airplane_Crashes['Fatalities'] = df_Airplane_Crashes['Fatalities'].astype(str).str.replace(r'\D', '', regex=True).astype(int)
```

####4.4 -Summary Text

```python
df_Airplane_Crashes['Summary'] = df_Airplane_Crashes['Summary'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

###5.4 - Airplane Crashes on Holidays vs. Non-Holidays
###5.5 - Top 5 Holidays with Most Airplane Crashes by Country

```python
import holidays
import matplotlib.pyplot as plt


countries = holidays.list_supported_countries()
global_holidays = holidays.HolidayBase()
country_holidays = {}
for country in countries:
  try: global_holidays.update(getattr(holidays, country)(years=range(1908, 2010)))
  except AttributeError:
    pass


for country in countries:
    try:
        country_holidays[country] = getattr(holidays, country)(years=range(1908, 2010))
    except AttributeError:
        pass

df_Airplane_Crashes['HolidayName'] = df_Airplane_Crashes['Date'].apply(
    lambda x: next((holiday_set.get(x) for country, holiday_set in country_holidays.items() if x in holiday_set), None)
)
df_Airplane_Crashes['HolidayCountry'] = df_Airplane_Crashes['Date'].apply(
    lambda x: next((country for country, holiday_set in country_holidays.items() if x in holiday_set), None)
)


holiday_crashes = df_Airplane_Crashes.dropna(subset=['HolidayName']).copy()
holiday_crashes['HolidayLabel'] = holiday_crashes.apply(lambda row: f"{row['HolidayName']} ({row['HolidayCountry']})", axis=1)
df_Airplane_Crashes['IsHoliday'] = df_Airplane_Crashes['Date'].apply(lambda x: x in global_holidays)

top_5_holidays = holiday_crashes['HolidayLabel'].value_counts().sort_values(ascending=False).head(5)
plt.figure(figsize=(6, 4))

holiday_crash_counts = df_Airplane_Crashes.groupby('IsHoliday').size()
holiday_crash_counts.plot(kind='bar', color=['blue', 'orange'])
plt.xticks(ticks=[0, 1], labels=['Non-Holiday', 'Holiday'], rotation=0)
plt.ylabel("Number of Crashes")
plt.title("Airplane Crashes on Holidays vs. Non-Holidays")
plt.show()

plt.figure(figsize=(8, 5))
top_5_holidays.plot(kind='bar', color='purple')
plt.xlabel("Holiday (Country)")
plt.ylabel("Number of Crashes")
plt.title("Top 5 Holidays with Most Airplane Crashes by Country")
plt.xticks(rotation=45, ha="right")
plt.show()
```

###5.6 -Crashes by Weather Conditions

```python
weather_keywords = ['storm', 'fog', 'rain', 'snow', 'lightning', 'wind', 'hail', 'turbulence', 'thunder', 'weather']


weather_fatalities = {}

for keyword in weather_keywords:
    condition_mask = df_Airplane_Crashes['Summary'].str.contains(keyword, case=False, na=False)
    weather_fatalities[keyword] = df_Airplane_Crashes.loc[condition_mask, 'Fatalities'].sum()

weather_fatalities_df = pd.DataFrame(list(weather_fatalities.items()), columns=['Weather Condition', 'Total Fatalities'])
print(weather_fatalities_df)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(weather_fatalities_df['Weather Condition'], weather_fatalities_df['Total Fatalities'], color='skyblue')
plt.xlabel("Weather Condition")
plt.ylabel("Total Fatalities")
plt.title("Fatalities in Weather-Related Crashes")
plt.xticks(rotation=45)
plt.show()
```

###5.7 - Is Flying Getting Safer Over Time?

```python
df_Airplane_Crashes['SurvivalRate'] = (df_Airplane_Crashes['Aboard'] - df_Airplane_Crashes['Fatalities']) / df_Airplane_Crashes['Aboard']
yearly_survival_rate = df_Airplane_Crashes.groupby('Year')['SurvivalRate'].mean()
df_Airplane_Crashes['Year'].value_counts().sort_index().plot(title='Crashes per Year')
yearly_survival_rate.plot(kind='line', color='green')
plt.title("Safety  Over Time (Survival Rate)")
plt.xlabel("Year")
plt.ylabel("Average Survival Rate")
plt.show()
```
